##Cifrado A

Crea la función cifrado_A que tome como parámetro una cadena de texto y la modifique desplazando cada letra 2 lugares. 

* Para no tener que trabajar con mayúsculas y minúsculas, la función debe convertir las mayúsculas en minúsculas con el metodo lower (). 

* Si no es una letra minúscula que deje el caracter como está.

* Utiliza una variable para almacenar el desplazamiento de las letras (será util para facilitar las funciones de los cifrados B y C).

**Pista:**Para hacer una lista con todas las letras del alfabeto puedes importar el paquete string y usar la función list().

      import string
      alfabeto=list(string.ascii_lowercase)

```python
#Crea la función del cifrado_A
import string
alfabeto=list(string.ascii_lowercase)
desplazamiento = 2
def cifrado_A(mensaje):

    mensaje_cifrado = ""
    for caracter in mensaje.lower():
        for indice, letra in enumerate(alfabeto):
            if caracter == letra:
                if indice + desplazamiento >= len(alfabeto):
                    # desborda!!!
                    llevar_a_cero = indice + desplazamiento - len(alfabeto)
                    caracter_cifrado = alfabeto[llevar_a_cero]
                else:
                    caracter_cifrado = alfabeto[indice +desplazamiento]
            elif caracter not in alfabeto:
                caracter_cifrado = caracter
        mensaje_cifrado += caracter_cifrado

    return mensaje_cifrado

cifrado_A('¡hola!')



```

##Enigma 

Una vez creados los tres cifrados, es hora de crear la función enigma. Recuerda que debe recibir una cadena de texto como parámetro y 3 claves de cifrado cuyo valor debe ser A,B o C.

El programa debe comprobar que las claves de cifrado  sean correctas y si no lo es pedir al usuario que introduzca un nuevo valor para dicha clave o claves hasta que sea correcto. **Pista:** Utiliza un bucle while.

Una vez comprobado debe realizar los tres cifrados en orden y la salida de ka fubción debe ser un diccionario con las claves (y el valor que las corresponda)

Por ejemplo:

Si la entada es:

* Mensaje a cifrar: '¡Felicidades! has creado un programa con 27 opciones distintas de cifrado'

* Clave de cifrado 1: C

* Clave de cifrado 2: B

* Cave de cifrado 3: A

La salida debe ser:

* {'clave_cifrado' : 'CBA', 'mensaje_cifrado' : '¡jipmgmhehiw! lew gviehs yr tvskveqe gsr 27 stgmsriw hmwxmrxew hi gmjvehs' }


```python
#Crea la función enigma, recuerda que debe devolver un diccionario con el mensaje cifrado y la clave.

def enigma(mensaje, clave_1, clave_2, clave_3):

    opciones = ["A", "B", "C"]
    while clave_1 not in opciones:
        clave_1 = str(input("Introduzca una clave_1 válida: A, B o C: "))
    while clave_2 not in opciones:
        clave_2 = str(input("Introduzca una clave_2 válida: A, B o C: "))
    while clave_2 not in opciones:
        clave_3 = str(input("Introduzca una clave_3 válida: A, B o C: "))

    clave = clave_1 + clave_2 + clave_3
    mensaje_cifrado = str(mensaje)
    clave_cifrado = {}

    for i in clave:
        if i == "A":
            mensaje_cifrado = cifrado_A(mensaje_cifrado)
        elif i == "B":
            mensaje_cifrado = cifrado_B(mensaje_cifrado)
        else:
            mensaje_cifrado = cifrado_C(mensaje_cifrado)

    clave_cifrado["clave_cifrado"] = clave
    clave_cifrado["mensaje_cifrado"] = mensaje_cifrado

    return clave_cifrado

str(enigma('Hola','C','B','A'))


```
